```python
import scipy.signal as sig
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
import pandas as pd
import scipy.optimize as so
import os
```

```python
nhead = 12
toVolt = 39.3216/((2**16) -1)
touAmp= (77/55)*toVolt
```

```python
VA=[1.790,1.794,1.79,1.807]
for i in range(4):
    file_path = r"C:\Users\n1cos\OneDrive\Documentos\Nucleus\scsi.22\SQ190_"+ str(i + 1) +"_noise_PTon.bin"
    #print("squidchar_noise_186" + str(i+1)+"M")
    fs=100e3
    nperseg=int(0.5*fs)
    a=VA[i]
    if os.path.isfile(file_path):
        #readRDT(file_path)
        fd = open(file_path, 'rb')
        read_data = np.fromfile(file=fd, dtype=np.int16)
        fd.close()
        read_data= toVolt*read_data[12:]/a
        f, pxx = sig.welch(read_data, fs=fs, nperseg=nperseg)
        f = f[:-1]
        ppx = 1e6 * np.sqrt(pxx[:-1])
        P10 = np.mean(ppx[(f >= 7) & (f <= 11)])
        P103 = np.mean(ppx[(f >= 900) & (f <= 1100)])
        print(P10)
        plt.plot(f, ppx,label=f"SQUID_"+ str(i + 1))
        
    else:
        print("No data taken")
#plt.title("QUAD 185",fontsize= 16)
#plt.axhline(y=4, color='r', linestyle='--', linewidth=5,label=" Shunt Noise")
file_path = r"C:\Users\n1cos\Downloads\clusterCryostatRun18\noise\noise_SQ190-3_000_PTON.bin"
fs=100e3
nperseg=int(0.5*fs)
    
os.path.isfile(file_path)
#readRDT(file_path)
fd = open(file_path, 'rb')
read_data = np.fromfile(file=fd, dtype=np.int32)
fd.close()
read_data= 39.3216/((2**24) -1)*read_data[12:]/1.70
f, pxx = sig.welch(read_data, fs=fs, nperseg=nperseg)
f = f[:-1]
ppx = 1e6 * np.sqrt(pxx[:-1])
plt.plot(f, ppx,label=f"SQUID_3")
plt.legend()
plt.title("ClusterCryostatRun19\ Noise\SQUID 190 ")
plt.yscale("log")
plt.xscale("log")
plt.grid(True)
plt.xlabel(r'$f$ [Hz]', fontsize= 16)
plt.ylabel(r'$ \frac{p A}{\sqrt{Hz}}$', fontsize= 16)
plt.show()        
```
