```python
import matplotlib.pyplot as plt

# Provided data
GAP = [5, 11, 20, 28, 51.3, 80.5, 96.5, 117.5, 139.5, 150]
xoff = -3.224+0.029
yoff = -5.091+1.846
zoff = -5.629+2.378
x_in_V = [-3.187, -3.548, -3.923, -4.119, -4.310, -4.350, -4.200, -4.053, -3.927, -3.964]
y_in_V = [-4.875, -4.911, -5.019, -5.136, -5.523, -6.194, -6.474, -6.806, -7.085, -7.195]
z_in_V = [-5.937, -5.952, -5.918, -5.857, -5.734, -5.377, -5.290, -5.230, -5.207, -5.171]

# Conversion to T (microteslas)
x_in_uB = [(x - xoff) * 10 for x in x_in_V]
y_in_uB = [(y - yoff) * 10 for y in y_in_V]
z_in_uB = [(z - zoff) * 10 for z in z_in_V]

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(GAP, x_in_uB, 'o-', label='Magnetic field in x direction')
plt.plot(GAP, y_in_uB, 'o-', label='Magnetic field in y direction')
plt.plot(GAP, z_in_uB, 'o-', label='Magnetic field in z direction')

# Adding labels and title
#plt.title("Magnetic field", fontsize=18)
plt.xlabel(r'$r$ [cm]', fontsize=16)
plt.ylabel(r'$B$ in [$\mu$T]', fontsize=16)

# Adding legend and grid
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Define grid
x, y, z = np.meshgrid(np.linspace(-40, 40, 4),
                      np.linspace(-40, 40, 4),
                      np.linspace(-40, 40, 4))

# Define vector components
u = 7.6 * x
v = 23.2 * y
w = -7.7 * z

# Create a figure with subplots
fig = plt.figure(figsize=(18, 12))

# 3D Vector Field Plot
ax1 = fig.add_subplot(141, projection='3d')
ax1.quiver(x, y, z, u, v, w, length=0.4, normalize=True)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
ax1.set_title('3D Vector Field')

# Plot x-y plane
ax2 = fig.add_subplot(142)
ax2.quiver(x[:, :, 0], y[:, :, 0], u[:, :, 0], v[:, :, 0], angles='xy', scale_units='xy', scale=1)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('x-y Plane')

# Plot x-z plane
ax3 = fig.add_subplot(143)
ax3.quiver(x[:, 0, :], z[:, 0, :], u[:, 0, :], w[:, 0, :], angles='xy', scale_units='xy', scale=1)
ax3.set_xlabel('X')
ax3.set_ylabel('Z')
ax3.set_title('x-z Plane')

# Plot y-z plane
ax4 = fig.add_subplot(144)
ax4.quiver(y[0, :, :], z[0, :, :], v[0, :, :], w[0, :, :], angles='xy', scale_units='xy', scale=1)
ax4.set_xlabel('Y')
ax4.set_ylabel('Z')
ax4.set_title('y-z Plane')

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```
